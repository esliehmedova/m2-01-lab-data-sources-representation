```python
valid_entries = []
invalid_entries = [] 

def validate_registrations(data):
    for entry in data:
        if (entry["username"] != "") and ("@" in entry["email"]) and (isinstance(entry["age"], int)):
            valid_entries.append(entry)
        else:
            invalid_entries.append(entry)
    return valid_entries, invalid_entries

print(validate_registrations(dataA)) 
```

```python
import pandas as pd 
import random
import numpy as np
```

```python
import os
```

```python
ride_type = ["standard", "premium", "pool"]
city = ["Berlin", "Seoul", "Nairobi", "Toronto", "Lima"] 
timestamps = pd.date_range(start="2025-01-01", periods=500, freq="min")


def generator(n):
    for i in range(1, n+1):
        driver_id = random.randint(1000, 1050)
        distance_km = random.uniform(1.0, 30.0) 
        fare_usd = random.uniform(5.0, 75.0) 
        
        yield {
            "ride_id" : i,
            "driver_id" : driver_id,
            "distance_km" : distance_km,
            "fare_usd" : fare_usd,
            "ride_type" : random.choice(ride_type),
            "city" : random.choice(city), 
            "timestamp" : timestamps[i-1]
        }
rides = list(generator(500)) 
```
